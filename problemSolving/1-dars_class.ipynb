```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def show_balance(self):
        print(f"Hozirgi balans: {self.balance}")

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} qo'shildi. Yangi balans: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Yetarli mablag' yo'q!")
        else:
            self.balance -= amount
            print(f"{amount} yechildi. Yangi balans: {self.balance}")
```

```python
# account yaratamiz
account = BankAccount(1000)
```

```python
# Hozirgi holat
account.show_balance()
```

```python
# Pul qo‘shamiz
account.deposit(500)
```

```python
# Pul yechamiz
account.withdraw(300)
```

```python
class StudentGrades:
    def __init__(self):
        self.students = {}

    def add_student(self, name, grades):
        self.students[name] = grades

    def average_score(self, grades):
        return sum(grades) / len(grades)

    def get_top_student(self):
        top_student = max(self.students, key=lambda x: self.average_score(self.students[x]))
        return top_student, self.average_score(self.students[top_student])

    def get_lowest_student(self):
        low_student = min(self.students, key=lambda x: self.average_score(self.students[x]))
        return low_student, self.average_score(self.students[low_student])

    def class_average(self):
        total = 0
        count = 0
        for grades in self.students.values():
            total += sum(grades)
            count += len(grades)
        return total / count
```

```python
sg = StudentGrades()
```

```python
sg.add_student("Ali", [90, 85, 88])
print("Eng yuqori:", sg.get_top_student())
```

```python
sg.add_student("Vali", [70, 75, 72])
print("Eng past:", sg.get_lowest_student())
```

```python
sg.add_student("Hasan", [95, 92, 96])
```

```python


print("O‘rtacha ball:", sg.class_average())
```

```python
class MyProfile:
    def __init__(self, name, age, skills, last_month_skills):
        self.name = name
        self.age = age
        self.skills = skills
        self.last_month_skills = last_month_skills

    def show_info(self):
        print(f"Ism: {self.name}")
        print(f"Yosh: {self.age}")
        print(f"Hozirgi skilllar: {', '.join(self.skills)}")

    def new_skills(self):
        new = set(self.skills) - set(self.last_month_skills)
        return list(new)

    def removed_skills(self):
        removed = set(self.last_month_skills) - set(self.skills)
        return list(removed)

    def compare(self):
        print(" Otgan oy bilan solishtirish:")
        print(f"Yangi organilgan skilllar: {self.new_skills()}")
        print(f"Yoqolgan skilllar: {self.removed_skills()}")
```

```python
last_month = ["HTML", "CSS", "Python"]
current = ["HTML", "CSS", "Python", "Machine Learning", "Git"]

profile = MyProfile("John", 31, current, last_month)

profile.show_info()
profile.compare()
```
